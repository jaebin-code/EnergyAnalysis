```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap
import seaborn as sns
import pandas as pd
```

```python
#  
df = pd.read_csv('heatmap.csv')


#    
start_row = 0
end_row = 334
data1 = df.iloc[start_row:end_row+1]

data1.to_numpy()
data1=np.array(data1)

#   0 
processed_data = np.zeros((31, 24), dtype=np.float)
#   
column1=data1[:,0]
column2=data1[:,1]
column3=data1[:,2]
column3=np.abs(column3)
for i in range(len(column1)):
    processed_data[int(column1[i])-1, int(column2[i])] = column3[i]
    
processed_data=np.transpose(processed_data)

#   
x = np.arange(0, 24)
y = np.arange(1, 32)

#   
colors = [(1, 1, 1), (0, 0.7, 1)]  #   RGB  
cmap = LinearSegmentedColormap.from_list('custom_cmap', colors)

#  
fig, ax = plt.subplots(figsize=(10, 8))
im = ax.pcolormesh(x, y-1, np.transpose(processed_data), cmap=cmap, edgecolors='black', linewidths=0.5)

#  
ax.set_xticks(np.arange(len(x)))
ax.set_yticks(np.arange(len(y)))
ax.set_xticklabels(x, fontsize=10)
ax.set_yticklabels(y, fontsize=10)

#    
plt.setp(ax.get_xticklabels(), rotation=45, ha="right", rotation_mode="anchor")

#   
plt.xlabel('Time', fontsize=12)
plt.ylabel('Date', fontsize=12)

# 
cbar = ax.figure.colorbar(im, ax=ax)

#  
plt.title('6th', fontsize=14)

#  
plt.tight_layout()
plt.show()
```
