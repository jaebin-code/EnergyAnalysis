# 계통 1

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import colors

# 데이터 가져오기
df = pd.read_csv('data1.csv')

# 특정 범위의 행 선택
start_row = 0
end_row = 334
data1 = df.iloc[start_row:end_row+1]

data1.to_numpy()
data1 = np.array(data1)

# 모든 값 0으로 초기화
processed_data = np.zeros((31, 24), dtype=np.float)
# 언패킹하여 열 분리
column1 = data1[:, 0]
column2 = data1[:, 1]
column3 = data1[:, 2]+1

for i in range(len(column1)):
    processed_data[int(column1[i]) - 1, int(column2[i])] = int(column3[i])

processed_data = np.transpose(processed_data)

# 데이터 크기 설정
x = np.arange(0, 24)
y = np.arange(1, 32)

colors = [
    (1,1,1),  # 흰색
    (0.984, 0.851, 0.659),  # 연한 주황색
    (0.886, 0.957, 0.733),  # 연한 녹색
    (0.827, 0.784, 0.980)   # 연한 보라색
]

cmap = LinearSegmentedColormap.from_list('pastel_cmap', colors, N=4)

# 히트맵 생성
fig, ax = plt.subplots(figsize=(10, 8))
im = ax.pcolormesh(x, y - 1, np.transpose(processed_data), cmap=cmap, vmin=0, vmax=3, edgecolors='black', linewidths=0.5)

# 컬러바 설정
cbar = ax.figure.colorbar(im, ax=ax, ticks=np.arange(0, 4))
cbar.ax.set_yticklabels(['Label 0', 'Label 1', 'Label 2', 'Label 3'])  # 라벨 표시

# 축 설정
ax.set_xticks(np.arange(len(x)))
ax.set_yticks(np.arange(len(y)))
ax.set_xticklabels(x, fontsize=10)
ax.set_yticklabels(y, fontsize=10)

# 눈금 및 라벨 회전
plt.setp(ax.get_xticklabels(), rotation=45, ha="right", rotation_mode="anchor")

# 축 레이블 설정
plt.xlabel('Time', fontsize=12)
plt.ylabel('Date', fontsize=12)


# 제목 설정
plt.title('6th', fontsize=14)

# 그래프 출력
plt.tight_layout()
plt.show()

```
